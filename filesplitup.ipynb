```python

# Load your data (assuming it's from a CSV for demonstration purposes)
#df = pl.read_csv('your_data.csv')

# Group by a column (replace 'group_column' with the actual column name)
group_column = "ticker"

# Define the path where files will be stored
output_dir = "option_data"
os.makedirs(output_dir, exist_ok=True)  # Ensure directory exists

unique_groups = df[group_column].unique()

# Iterate through each group and save to CSV
for group_value in unique_groups:
    # Filter the DataFrame for the current group
    group_df = df.filter(pl.col(group_column) == group_value)
    
    # Save the group DataFrame to a CSV file
    file_name = f"{output_dir}/{group_value}_option_data.csv"
    group_df.write_csv(file_name)
```
